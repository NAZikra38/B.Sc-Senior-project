Imagine a particle (like an electron) trapped in a box. This box has walls at x = -a and x = +a. The particle cannot escape — this is the infinite square well.
Now, inside this box, someone places a small speed bump (a potential barrier, V_0) in the middle. This speed bump can grow taller or shrink as time passes.
Your job is to watch how the particle's probability density changes as the barrier grows/shrinks — in 1D, 2D, and 3D animations.


**The Box (Infinite Square Well):-**
1. The particle lives between x = -a and x = +a
2. Outside the box, the probability of finding the particle is zero (infinite walls).**bold text**
3. Inside the box, the particle has certain allowed "shapes" it can take — these are called eigenstates ψ₁, ψ₂, ψ₃...
4. Think of them like vibration modes of a guitar string — ψ₁ has one hump, ψ₂ has two humps, etc.

**The Initial Wavefunction ψ** :-

The particle doesn't start in a pure state. It starts as a mixture of the first two shapes (ψ₁ and ψ₂):
   
$$\\ψ=(\frac{(1−α)ψ_1+αψ_2}{\sqrt{1+2α^2+2α}}$$

1. α controls the mixing. If α = 0, you have pure ψ₁. If α = 1, you have pure ψ₂
2. The bottom part (denominator) is just a normalization factor — it ensures the total probability adds up to 1.

(It can be imagined like mixing two paint colors — α controls how much of each color you use.)


**The Barrier V(x, t)**

Inside the well, there is a small barrier:-

$$\\ V(a)=θ(t)⋅V_0​⋅[θ(x+a)(x−a)]$$
1. θ(t) is a "switch" — it turns the barrier on/off or ramps it up smoothly with time
2. V₀ is the height of the barrier — you can change this in the code with a slider
3. The barrier only exists inside the well (between -a and +a)
4. The three conditions tell us: i. V(x) < 0 → you're deep in the well (particle is free here); ii. V(x) ≈ 0 → you're at the barrier (the particle "feels" it); iii. V(x) > 0→ outside the well (impossible to be here due to infinite walls)

# **Mathematical Foundation**

The split-step method solves the Schrodinger equation:

$$ iℏ \frac{∂ψ}{ ∂t} = \hat{H}ψ = (\hat{T}+ \hat{V}) ψ $$

By approximating the evolution operator:

$$ e^{(−i \frac{\hat{H}dt}{ℏ})} ≈ e^{(−i \frac{\hat{V}dt}{2ℏ})}e^{(−i \frac{\hat{T}dt}{ℏ})}e^{(−i \frac{\hat{V}dt}{2ℏ})}+ O(dt^3) $$

This method is unitary (preserves normalization) and second-order accurate.

## **3 Computational Strategy**

**Solution Strategy**
---
1. Eigenstate Calculation:

• Use particle-in-a-box solutions as basis states\
• $ψ_1(x)$: ground state (n=1)\
• $ψ_2(x)$: first excited state (n=2)\
• Normalize eigenstates: $\int_{1}^{-1}|ψ_{n}|^2 dx = 1 $

2. Initial Condition:\
• Superposition: $ψ(t = 0) = \\ψ=(\frac{(1−α)ψ_1+αψ_2}{\sqrt{1+α^2+2α}}$ \
• Parameter $α = 0.3$ controls state mixing \
• Ensures proper normalization

3. Time Evolution:\
• Split-step Fourier method (2nd order accurate) \
• Kinetic operator in k-space: $exp(−iℏk^2dt/(2m))$ \
• Potential operator in x-space: $exp(−iV (x,t)dt/ℏ)$ \
• Unitary evolution preserves normalization

4. Potential Barrier:\
• Rectangular barrier: $V (x,t) = α(t)V_0$ for $x ∈ [−a,a]$ \
• Time modulation: $ α(t) = 0.5+0.5sin(2πt/T)$ \
• Varies smoothly from $0$ to $V_0$

5. Dimensional Extension: \
• Separable solutions: $ψ_nD = \prod_{i=1}^{n} ψ_{1D}(x_i) $ \
• 2D: $ψ(x,y) = ψ_{1D}(x)ψ_{1D}(y) $ \
• 3D: $ψ(x,y,z) = ψ_{1D}(x)ψ_{1D}(y)ψ_{1D}(z)$

6. Visualization:\
• 1D: Line plot of $|ψ|^2$ vs $x$ \
• 2D: Contour plot of $|ψ(x,y)|^2$ \
• 3D: Isosurface rendering (threshold = 30% of max) \
• Color coding: Blue for $ψ_1$ component, Red for $ψ_2$ component

7. Performance Optimization:\
• 1D: 512 grid points (high resolution) \
• 2D: 128×128 grid points (moderate resolution) \
• 3D: 64×64×64 grid points (lower resolution for performance) \
• Pre-compute all time steps before rendering

$$ iℏ \frac{∂ψ}{∂t} = \hat{H}ψ = −\frac{ℏ^2}{2m} \frac{∂^2}{∂x^2} + V(x,t) ψ $$

$$ ψ(x,0) = \frac{(1−α)ψ_1(x)+αψ_2(x)}{\sqrt{1 +α^2 −2α}} $$

\end{cases} $$

$$ρ(x,t) = |ψ(x,t)|^2 = ψ^∗(x,t)ψ(x,t) $$

 $$∫_{-∞}^{∞}
|ψ(x,t)|^2dx = 1 $$
for all the value of t\

$$ c_n(t) = ⟨ψ_n|ψ(t)⟩ = ∫_{-1}^{1}
ψ^∗
